# 파이썬 리스트 vs 넘파이 리스트

* 동적 타입의 파이썬 리스트 vs 고정 타입의 넘파이 배열

```python
"""파이썬 리스트 vs 넘파이 리스트"""
import numpy as np
np.__version__  #버전 확인

L1 = list(range(10))
print(type(L1[0]))

L2 = list(str(i) for i in L1)
print(type(L2[0]))

L3 = [True, "A", 3.0, 4]
print([type(k) for k in L3])  
#서로 다른 데이터 타입을 담는 파이썬 리스트의 유연성. 그러나 이러한 유연성에서는 비용이 따른다
#동적타입의 리스트(파이썬 리스트) vs 고정 타입의 리스트 (NumPy 리스트)
#동적 타입의 리스트는 더 유연할지 몰라도 데이터의 저장과 가공에는 NumPy 스타일이 훨씬 효율적이다

ll = np.array(L3)
print(ll)
print([type(k) for k in ll])  #모든 원소들이 문자열로 바뀜
```

# 넘파이 배열 속성 지정

# 배열 인덱싱 : 단일 요소에 접근하기

```python
"""배열 인덱싱 : 단일 요소에 접근하기"""
x1 = np.random.randint(10, size = 6)         #1차원 배열 생성
x2 = np.random.randint(10, size = (3,4))     #2차원 배열 생성

#일차원 배열 인덱싱은 파이썬 리스트와 같으므로 생략

print(1, x2)
print(2, x2[0,0])   #2차원 인덱싱은 [행,열] 순서
print(3, x2[2,0])
print(4, x2[2,-1])

x2[0,0] = 500   #인덱싱을 이용해 값의 수정도 가능하다
print(5, x2)       #주의할 점: NumPy 배열은 고정 타입을 가진다
                ######## -> 정수 배열에 부동 소수점 값을 입력하려고 하면 소수점 이하를 아무말 없이 잘라버릴 것

x1[0] = 3.141592
print(6, x1)
```

# 배열 슬라이싱 : 하위 배열에 접근하기

  * 슬라이싱으로 생성된 하위 배열은 원본의 사본(copy)가 아닌 뷰(view)이다.

  * 사본을 만들어주는 .copy() 메서드가 있음

```python
"""배열 슬라이싱 : 하위 배열에 접근하기"""
#       x[ start : stop : step ]

#1차원 배열 슬라이싱은 파이썬 리스트와 같으므로 생략
## step이 음수인 경우 배열을 거꾸로 만드는 편리한 방법이 될 수 있다

#2차원
x2 = np.random.randint(10, size = (3,4))     #2차원 배열 생성
print(1, x2)

print(2, x2[:2, :3])   #2개의 행, 3개의 열

print(3, x2[:3, ::2])  #모든 행, 한 열 걸러 하나씩

print(4, x2[::-1, ::-1])  #역으로 변환

print(5, x2[:, 0])        #배열의 단일 행이나 열에 접근할 때 : 슬라이싱과 인덱싱 동시에 사용
print(6, x2[0, :])

print("#########################")
#사본이 아닌 뷰로서의 하위 배열
#NumPy 배열 슬라이싱은 배열 데이터의 사본(copy)이 아니라 뷰(view)를 제공한다는 것
print(x2)
x2_sub = x2[:2, :2]   #2*2 하위 배열을 추출, x2_sub으로 저장

x2_sub[0,0] = 999     #하위 배열 데이터 수정
print(x2)             #원본 데이터도 수정됨을 확인

x2_sub_copy = x2[:2,:2].copy()    #copy() 메서드로 복사 가능
x2_sub_copy[0:0] = 50000
print(x2)   #원본 데이터 바뀌지 않음을 확인
```

# 배열 연결 및 분할

  * np.concatenate()
    - axis 매개변수 : 0번째, 1번째, 2번째 ... 축을 따라서 연결 (축 개수 세는 방향으로)

  * np.vstack() , np.hstack() : 복합 차원들의 배열 연결


---

  * np.split() : 괄호 안에 분할 지점(인덱스)를 전달

  * np.vsplit() , np.hsplit()


```python
"""배열 연결 및 분할"""
##배열 연결
x = np.array([1,2,3])
y = np.array([4,5,6])
xy = np.concatenate((x, y))   #배열을 결합하는 concatenate 함수
print(1, xy)

z = [98, 99, 100]         
xyz = np.concatenate((x,y,z)) #한번에 2개 이상의 리스트들 결합 가능
print(2, xyz)

##2차원 배열 연결
grid = np.array([[1,2,3],
                 [4,5,6]])
L1 = np.concatenate((grid, grid), axis = 0) #axis 매개변수 0 : 첫번째 축을 따라 연결
print(3, L1)
L2 = np.concatenate((grid, grid), axis = 1) #axis = 1 : 두번째 축을 따라 연결
print(4, L2)

#혼합 차원의 배열로 작업할 때는 np.vtack, np.hstack 함수를 사용하는것이 더 명확하다
x = np.array([1,2,3])
grid = np.array([[9,8,7],
                 [6,5,4]])

Vertical = np.vstack([x, grid])   #Vertical Stack : 수직 스택
print(5, Vertical)

y = np.array([[99],
              [99]])
Horizontal = np.hstack([y, grid]) #Horizontal Stack : 수평 스택
print(6, Horizontal)

#마찬가지로 np.dstack 은 세번째 축을 따라 배열을 쌓을 것이다 (Dimensional Stack, 차원 스택)



###############################
#배열 분할하기
##결합의 반대는 분할이다. np.split, np.hsplit, np.vsplit 함수로 구현
x = [0,1,2,3,4,5,6,7]
x1, x2, x3 = np.split(x, [3,5])   
#분할 지점을 알려주는 인덱스 목록을 전달하자. N개의 분할 지점은 N+1개의 하위배열을 만든다
print(7, x1, x2, x3)


grid = np.arange(0,16).reshape((4,4))
upper, lower = np.vsplit(grid, [2])
print(8)
print(upper)
print(lower)

left, right = np.hsplit(grid, [2])
print(9)
print(left)
print(right)

#비슷하게 np.dsplit 은 세번째 축을 따라 배열을 분할한 것이다 (Dimensional Split)
```

# 유니버설 함수 소개 (Universal Functions, UFuncs)

* Universal Functions  vs  파이썬 반복문

  -> 유니버설 함수가 더욱 빠르고 효율적임!

  -> 특히 배열의 크기가 커질수록 그 차이가 확연하다.

```python
"""UFuncs 소개"""
#파이썬의 루프(반복문)은 느리다!!!!
#NumPy의 벡터화 연산 : NumPy 배열의 값에 반복된 연산을 빠르게 수행하는 것이 주된 목적인 ufuncs를 통해 구현

bunja = np.ones(5)
bunmo = np.array([1,2,4,5,10])
print(bunja/bunmo)              #두 배열간의 연산도 가능케 함

x = np.arange(9).reshape(3,3)
print(x**2)     #다차원 배열에서도 가능하다!

print(1/bunmo)  #스칼라와 배열의 계산도 가능
```

# NumPy 유니버설 함수(UFuncs)

* 이하 코드에 언급되지 않았지만 쌍곡선 삼각함수(하이퍼볼릭 삼각함수), 비트 연산, 비교연산자, 라디안을 각도로 변환, 반올림과 나머지 등등 훨씬 더 많은 UFunc가 있다. 

  * NumPy 문서에서 더 많은 기능을 확인해보자

* scipy.special 모듈에서 더욱 많은 함수들 구현 가능

```python
"""NumPy 유니버설 함수(UFuncs)"""
# 배열 산술 연산 : 파이썬 기본 연산자 사용 가능, +각각 대응되는 ufuncs들이 존재한다
np.add(2,3)           #덧셈
np.subtract(2,3)      #뺄셈
np.negative(2)        #단항 음수
np.multiply(2,4)      #곱셈
np.divide(2,4)        #나눗셈
np.power(2,3)         #지수 연산 (거듭제곱)
np.floor_divide(3,2)  #바닥 나눗셈 (몫 출력)
np.mod(3,2)           #나머지 연산

#절댓값 함수
x = np.array([-2,-1,0,1,2])
np.abs(x)         #절댓값 함수 np.absolute, np.abs라는 별칭으로도 사용가능

#삼각함수
theta = np.linspace(0, np.pi, 3)    #먼저 각도 배열을 정의하자 : (0, pi/2, pi)

print("theta: ", theta)
print("sin(theta): ", np.sin(theta))
print("cos(theta): ", np.cos(theta))
print("tan(theta): ", np.tan(theta))
print(" ")

x = [-1, 0, 1]

print("x:", x)
print("arcsin(x): ", np.arcsin(x))  #역삼각함수도 된다!
print("arccos(x): ", np.arccos(x))
print("arctan(x): ", np.arctan(x))
print(" ")

#지수와 로그
y = [1,2,3]

print("y : ", y)
print("e^y : ", np.exp(y))
print("2^y : ", np.exp2(y))
print("3^y : ", np.power(3,y))
print(" ")

z = [1,2,4,10]
print("z : ", z)
print("ln(z) : ", np.log(z))
print("log2(z) : ", np.log2(z))
print("log10(z) : ", np.log10(z))
print(" ")

####### x가 매우 작은 경우 np.log 나 np.exp 보다 더 정확한 값을 위한 np.log1p 나 np.expm1 도 있다

#특화된 유니버설 함수
from scipy import special       #scipy.special 모듈에 훨씬 많은 수학적 함수들이 있다
a = [1,5,10]

print("a : ", a)
print("Gamma(a) : ", special.gamma(a))          #감마함수과 그 관련 함수
print("ln{Gamma(a)} : ", special.gammaln(a))
print("Beta(a, 2) : ", special.beta(a,2))
print(" ")

b = np.array([0, 0.3, 0.7, 1.0])
print("b : ", b)
print("erf(b) : ", special.erf(b))              #오차함수(가우스 적분), 그 보수와 역수
print("erfc(b) : ", special.erfc(b)) 
print("erfinv(b) : ", special.erfinv(b)) 
```

# 브로드캐스팅 규칙

* NumPy의 브로드캐스팅은 두 배열 사이의 상호작용을 결정하기 위해 엄격한 규칙을 따른다
  1. 두 배열의 차원 수가 다르면 더 작은 수의 차원을 가진 배열 형상의 앞쪽(왼쪽)을 1로 채운다
  2. 두 배열의 형상이 어떤 차원에서도 일치하지 않는다면 해당 차원의 형상이 1인 배열이 다른 형상과 일치하도록 늘어난다
  3. 임의의 차원에서 크기가 일치하지 않고 1도 아니라면 오류가 발생한다

```python
"""브로드캐스팅 예제 1"""
M = np.ones((2,3))  #2차원
a = np.arange(3)    #1차원
print(M+a)
```

```python
"""브로드캐스팅 예제 2"""   #두 배열 모두 브로드캐스팅이 필요한 예제
a = np.arange(3).reshape((3,1))
b = np.arange(3)
print(a+b)
```

```python
"""브로드캐스팅 예제 3"""   #두 배열이 서로 호환되지 않는 경우의 예제
M = np.ones((3,2))
a = np.arange(3)
print(M+a)        #에러 발생 !
```

* 팬시 인덱싱을 사용하면 반환값은 인덱싱 대상 배열의 형상이 아닌 **"브로드캐스팅 된 인덱스의 형상"**을 반영한다

# 예제: K-최근접 이웃 알고리즘

```python
"""부분 정렬: 파티션 나누기"""

#2차원 평면에 임의의 점 10개를 가지는 집합을 만들자
rand = np.random.RandomState(42)
X = rand.rand(10,2)
print(X)

#시각화
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
plt.scatter(X[:, 0], X[:, 1], s = 100)
plt.show()

#각 쌍의 점 사이의 거리 계산
dist_sq = np.sum((X[:, np.newaxis, :] - X[np.newaxis,:,:])**2, axis = 2)

#대각선이 모두 0인지 확인하기 (각 점과 자기 자신 사이의 거리 집합)
dist_sq.diagonal()

#제곱 거리를 변환
nearest = np.argsort(dist_sq, axis = 1)
print(nearest)

K = 2
nearest_partition = np.argpartition(dist_sq, K+1, axis = 1)

plt.scatter(X[:,0], X[:,1], s = 100)

#각 점을 두 개의 가장 가까운 이웃과 선으로 이음
for i in range(X.shape[0]):
  for j in nearest_partition[i, :K+1]:
    #X[i]부터 X[j]까지 선으로 이음
    #zip 매직 함수를 이용함
    plt.plot(*zip(X[j], X[i]), color = "black")

```

# 구조화된 배열 만들기

```python
"""구조화된 배열 만들기"""
#복합적인 이종 데이터의 효율적인 저장을 위한 구조화된 배열

name = ["Alice", "Bob", "Cathy", "Doug"]
age = [25, 45, 37, 19]
weight = [55.0, 85.5, 68.0, 61.5]

##구조화된 배열을 위해 복합 데이터 타입 사용
data = np.zeros(4, dtype = {"names":("name", "age", "weight"),
                            "formats":("U10", "i4", "f8")})     #빈 컨테이너 배열 생성

data["name"] = name
data["age"] = age
data["weight"] = weight   #데이터 채워넣기
print(data)

#값을 인덱스나 이름으로 참조할 수 있는 점이 편리함
#부울리언 인덱싱으로 필터 적용과 같이 복잡한 연산도 가능하다

#여러 가지 지정 방법이 존재한다 --> 교재 참고 p108
```
